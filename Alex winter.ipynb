```python
import cv2
import numpy as np
import scipy
```

```python
image = cv2.imread('gambar11.jpg')
```

```python
def Winter(img):
    increaseLookupTable = LookupTable([0, 64, 128, 256],
                                      [0, 80, 160, 256])
    decreaseLookupTable = LookupTable([0, 64, 128, 256],
                                      [0, 50, 100, 256])
    blue_channel, green_channel,red_channel = cv2.split(img)
    red_channel = cv2.LUT(red_channel,
                          decreaseLookupTable).astype(np.uint8)
    blue_channel = cv2.LUT(blue_channel,
                           increaseLookupTable).astype(np.uint8)
    win= cv2.merge((blue_channel, green_channel, red_channel))
    return win
```

```python
import numpy as np

def LookupTable(x, y):
    # Create a lookup table for 0-255 pixel values using linear interpolation
    x_array = np.array(x)
    y_array = np.array(y)
    return np.interp(np.arange(256),
                     x_array, y_array).astype(np.uint8)

#making the winter img
a10 = Winter(image)
filename = 'winter11.jpg'
# Using cv2.imwrite() method
# Saving the image
cv2.imwrite(filename, a10)
```
