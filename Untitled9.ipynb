```python
import cv2
import numpy as np
import scipy
```

```python
image = cv2.imread("gambar.jpg")
```

```python
def gray(img):
  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  return gray
```

```python
a0=gray(image)
cv2.imwrite('sampel1.jpg',a0)
```

```python
def bright(img,beta_value):
  img_br = cv2.convertScaleAbs(img,beta=beta_value)
  return img_br
```

```python
a3= bright(image,-60)
```

```python
cv2.imwrite('sampel.jpg',a3)
```

```python
def invert(img):
  inv=cv2.bitwise_not(img)
  return inv
```

```python
a9=invert(image)
```

```python
cv2.imwrite('sampel2.jpg',a9)
```
