```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
gambar = cv2.imread("gambar.jpg")
plt.imshow(gambar)
```

```python
gambar1 = cv2.cvtColor(gambar, cv2.COLOR_RGB2XYZ)
plt.imshow(gambar1)
```

```python
cv2.imwrite("bas9.jpg", gambar1)
```

```python
img_a = cv2.imread('wa10.jpg')
plt.imshow(img_a)
```

```python
img_b = cv2.imread('bb.png')
img_b = cv2.cvtColor(img_b, cv2.COLOR_BGR2RGB)
plt.imshow(img_b)
```

```python
img_a = cv2.resize(img_a, (1080, 1080))
img_b = cv2.resize(img_b, (1080, 1080))
plt.imshow(img_b)
```

```python
blended = cv2.addWeighted(src1 = img_a, alpha = 0.7, src2 = img_b, beta = 0.3, gamma = 0 )
plt.imshow(blended)
```

```python
cv2.imwrite("wa20.jpg", blended)
```
