```python
import numpy as np
import math
import cv2
from matplotlib import pyplot as plt
from PIL import ImageStat
from PIL import Image
```

```python
imgori=cv2.imread('gambar1.jpg',-1)
img=cv2.imread('gambar1.jpg',0)
out_img=cv2.equalizeHist(img)
```

```python
plt.subplot(121),plt.imshow(imgori,cmap='gray'),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(out_img,cmap='gray'),plt.title('hinstogram')
plt.xticks([]),plt.yticks([])
plt.show()
cv2.imwrite('g2.jpg',out_img)
```
