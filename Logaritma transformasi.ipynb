```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import copy
import math
```

```python
imgori=cv2.imread('Gambar.jpg',-1)
img=cv2.imread('Gambar.jpg',0)
out_img=copy.copy(img)
c=255/math.log(np.max(img)+1.2)
```

```python
for i in range(0,out_img.shape[0]):
   for j in range(0,out_img.shape[1]):
       out_img[i,j]=c*math.log(out_img[i,j]+1.2)
```

```python
plt.subplot(121),plt.imshow(imgori,cmap='gray'),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(out_img,cmap='gray'),plt.title('log transformation')
plt.xticks([]),plt.yticks([])
plt.show()
cv2.imwrite('gambar33.jpg',out_img)
cv2.imwrite('img4.jpg',img)
```
