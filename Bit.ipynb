```python
import numpy as np
import cv2
import copy
from matplotlib import pyplot as plt
import math
```

```python
imgori=cv2.imread('img.jpg',-1)
img=cv2.imread('img.jpg',0)
bit_plane_1=copy.copy(img)
```

```python
for i in range (0,img.shape[0]):
   for j in range (0,img.shape[1]):
      bit_plane_1[i,j]=img[i,j]&128
```

```python
plt.subplot(121),plt.imshow(imgori, cmap='gray'),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(bit_plane_1, cmap='gray'),plt.title('bit plane slicing')
plt.xticks([]),plt.yticks([])
plt.show()
cv2.imwrite('h.jpg',bit_plane_1)
```
