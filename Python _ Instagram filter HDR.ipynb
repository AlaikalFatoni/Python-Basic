```python
import cv2
import numpy as np
import scipy
```

```python
image = cv2.imread('gambar.jpg')
```

```python
def hdr(img):
  hdr = cv2.detailEnhance(img,sigma_r=0.15,sigma_s=12)
  return hdr
```

```python
a1 = hdr(image)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(a1)
cv2.imwrite('hdr16.jpg',a1)
```
