```python
import numpy as np
import math
import cv2
from matplotlib import pyplot as plt
```

```python
imgori=cv2.imread('gambar3.jpg',-1)
image=cv2.imread('gambar3.jpg',0)
a=int(image.max())
b=int(image.min())
x=0
y=255
c=image.astype(float)
```

```python
out_image=((c-b)*((y-x)/(a-b)))+x
out_image=out_image.astype(int)
```

```python
plt.subplot(121),plt.imshow(imgori,cmap='gray'),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(out_image,cmap='gray'),plt.title('contrast stretching')
plt.xticks([]),plt.yticks([])
plt.show()
cv2.imwrite('foto3.jpg',out_image)
cv2.imwrite('foto0.jpg',image)
```
