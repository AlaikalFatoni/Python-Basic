```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imgori = cv2.imread('gambar.jpg',-1)
img = cv2.imread('gambar.jpg',0)
img2 = 255 -img
```

```python
plt.subplot(121),plt.imshow(imgori,cmap='gray'),plt.title('original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(img2,cmap='gray'),plt.title('citra negatif')
plt.xticks([]),plt.yticks([])
plt.show()
cv2.imwrite('d.jpg',imgori)
plt.savefig('cn1.jpg')
```
