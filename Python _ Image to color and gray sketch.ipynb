```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def applypencilsketch(image, display=True):
  gray_sketch, color_sketch = cv2.pencilSketch(image,sigma_s=20,sigma_r=0.5,shade_factor=0.02)
  if display:
    plt.figure(figsize=[15,15]);
    plt.subplot(131);
    plt.imshow(image[:,:,::-1]);
    plt.title('ori');
    plt.axis('off');
    plt.subplot(132);
    plt.imshow(color_sketch[:,:,::-1]);
    plt.title("color sketch");
    plt.axis('off');
    plt.subplot(133);
    plt.imshow(gray_sketch,cmap='gray');
    plt.title("gray sketch");
    plt.axis('off');
  else:
    return color_sketch
```

```python
image = cv2.imread("gambar2.jpg")
applypencilsketch(image)
```
